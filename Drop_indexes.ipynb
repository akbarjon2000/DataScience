```
import pandas as pd
```

```
data = {
    "a":[1,2,3,4,4,5,6],
    "b":[8,7,6,8,9,6,6],
    "c":['s','f','g','r','e','e', 'g']
}
```

```
frame = pd.DataFrame(data)
```

```
frame
```

```
frame.drop(0)
```

```
frame.drop(0, axis = 0)
```

```
frame.drop(0, axis=0, inplace=True)
```

```
frame
```

```
frame.drop('a', axis = 1)
```

```
frame.drop('a', axis=1, inplace=True)
```
