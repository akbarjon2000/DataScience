```
def loss_function(k ,b, waste):
  total_error = 0
  for i in range(len(waste)):
    x = waste.iloc[i].i_day
    y = waste.iloc[i].m_food_wst_amt
    total_error += (y-(k*x + b)) ** 2
  total_error / float(len(waste))
```

```
def gradient_decent(m_now, b_now, waste, L):
  m_gradient = 0
  b_gradient = 0

  n = len(waste)

  for i in range(n):
    x = waste.iloc[i].i_day
    y = waste.iloc[i].m_food_wst_amt
#np.clip((y - (m_now * x + b_now)), -1e10, 1e10)
    m_gradient += -(2/n) * x * (y - (m_now * x + b_now))
    b_gradient += -(2/n) * (y - (m_now * x + b_now))
  m = m_now - m_gradient * L
  b = b_now - b_gradient * L
  return m, b
```
