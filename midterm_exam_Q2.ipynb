```python
import pandas as pd

# Read the CSV file and skip lines with parsing errors
file_path = "/Users/akbarjon/Downloads/waste.xlsx"
df = pd.read_excel(file_path, sheet_name="2017년")
df.head(10)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Load your dataset (replace 'your_data.csv' with your actual data file)
data = df

# Data Preparation for Waste Amount Prediction
X_amount = data[['m_food_wst_cnt', 'm_food_wst_g']]  # Features affecting Waste Amount
y_amount = data['m_food_wst_amt']  # Target variable (Waste Amount)

# Data Preparation for Waste Count Prediction
X_count = data[['m_food_wst_amt', 'm_food_wst_g']]  # Features affecting Waste Count
y_count = data['m_food_wst_cnt']  # Target variable (Waste Count)

# Split the data into training and test sets for both predictions
X_amount_train, X_amount_test, y_amount_train, y_amount_test = train_test_split(X_amount, y_amount, test_size=0.2, random_state=42)
X_count_train, X_count_test, y_count_train, y_count_test = train_test_split(X_count, y_count, test_size=0.2, random_state=42)

# Waste Amount Prediction using Linear Regression
model_amount = LinearRegression()
model_amount.fit(X_amount_train, y_amount_train)
y_amount_pred = model_amount.predict(X_amount_test)
mse_amount = mean_squared_error(y_amount_test, y_amount_pred)
print(f"Waste Amount Mean Squared Error: {mse_amount}")

# Waste Count Prediction using Random Forest Regression
model_count = RandomForestRegressor(n_estimators=100, random_state=42)
model_count.fit(X_count_train, y_count_train)
y_count_pred = model_count.predict(X_count_test)
mae_count = mean_absolute_error(y_count_test, y_count_pred)
print(f"Waste Count Mean Absolute Error: {mae_count}")
```
